```python
df.iloc[5:11].plot(kind = 'bar',figsize = (20,10)) 
# 5:11 = [5,6,7,8,9,10] , iloc can have two [] one for row and other for col like......iloc[5:11],[1:4]
# the kind arg is used to showing charts with diff forms by default its 'line'


# iloc is used to select and show only selected data(here form row 5 to 11)

#plot only shows bar or lines of numerical data not categorical data
```
